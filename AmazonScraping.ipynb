***Get Articles Link From the above text:: These Various classes and attrs that are used based on the Inspection of Web Page Thus they might differ for different web pages***

```python
for k in soup.find_all('div',attrs={'class':'blog-post-inr'}):
  for j in k.find_all('div',class_='col-md-8'):
    link_list=[]
    for m in j.find_all('h2',class_='post-title'):
      link_list.append(m.find('a')['href'])
```

***Get Text From Various Articles Links that are retrieved from the above Text Based on Inspection of 1 Web page ;Similar for all the other Web pages in the above link***

```python
doc_all_rank=[]
for link in link_list:
  doc_1=''
  linkdriver=driver.get(link)
  content=driver.page_source
  soup=BeautifulSoup(content)
  for k in soup.find_all('div',class_='post-body'):
    for m1 in k.find_all('p')[:]:
      doc_1+=m1.text
  doc_1+=str(link)
  doc_all_rank.append(doc_1)
```

```python
for link in link_list_doc2:
  doc_1=''
  linkdriver=driver.get(link)
  content=driver.page_source
  soup=BeautifulSoup(content)
  for k in soup.find_all('main',attrs={'id':'main'}):
    for j in k.find_all('div',class_='main_single'):
      for j1 in j.find('div',class_='post_image'):
        if j1.find('a')!=None:continue
        else: doc_1+=j1.text
      for j2 in j.find_all('div',class_='post_content'):
        for j22 in j2.find_all("p"):
          doc_1+=j22.get_text()
        for j22 in j2.find_all("h2"):
          doc_1+=j22.get_text()
        for j22 in j2.find_all("h3"):
          doc_1+=j22.get_text()
  doc_list_2.append(doc_1)
```
